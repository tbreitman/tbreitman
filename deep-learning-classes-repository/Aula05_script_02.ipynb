```python
import pandas as pd
```

```python
df = pd.DataFrame([x, y, z]).T
df.columns = ['ordem', 'valor', 'classe']
df.head()
```

```python
df.ordem
```

```python
df.classe.value_counts()
```

```python
df.classe.unique()
```

```python
df.classe.value_counts().keys()
```

```python
keys = [int(v) for v in df.classe.value_counts().keys()]
values = [int(v) for v in df.classe.value_counts().values]
keys, values
```

```python
plt.bar(keys, values, color=['b', 'k', 'c'], width=0.8)
plt.title('Frequncia classes')
plt.xticks(ticks=np.arange(min(df.classe.unique()), max(df.classe.unique()) + 1, 1))
plt.xlabel('Classes')
plt.ylabel('Frequncia')
# for rect, label in zip(keys, values):
#     height = rect.get_height()
#     plt.text(
#         rect.get_x() + rect.get_width() / 2, height + 5, label, ha="center", va="bottom"
#     )
# plt.grid()
# plt.ylim([0, 100])
# plt.ylim([0,100])
plt.grid()
plt.show()
```

```python
keys_col2 = [v for v in df.valor.value_counts().keys()]
values_col2 = [int(v) for v in df.valor.value_counts().values]
```

```python
keys_col2
```

```python
plt.bar(keys_col2, values_col2, color=['r', 'm', 'y'], width=0.8)
plt.title('Frequncia classes')
plt.xticks(ticks=np.arange(min(df.valor.unique()), max(df.valor.unique()) + 1, 1))
plt.xlabel('Classes')
plt.ylabel('Frequncia')
plt.grid()
plt.show()
```
